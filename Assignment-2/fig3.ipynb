```python
import numpy as np
import xarray as xr
import pandas as pd
import cartopy as cr
import cartopy.feature as cf
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

cold spell

```python
cold_spell = xr.open_dataset("cold_spell.nc")
```

```python
cold_spell = np.mean(cold_spell['cold_waves_per_time_period'],axis=(0))

lon = cold_spell.longitude; lat = cold_spell.latitude
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(cold_spell,extent=(lon.min(),lon.max(),lat.min(),lat.max()),cmap='cool',origin='upper')

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('Cold spell')
#plt.savefig("../Assignment-2/plots/fig3/cold spell spatial")
```
