```python
import numpy as np
import xarray as xr
import pandas as pd
import cartopy as cr
import cartopy.feature as cf
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

cold nights

```python
cold_nights = xr.open_dataset("cold_nights.nc")
cold_nights = np.mean(cold_nights['cold_nights_percent_wrt_10th_percentile_of_reference_period'],axis=(0))
```

```python
lon = cold_nights.longitude; lat = cold_nights.latitude
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(cold_nights,extent=(lon.min(),lon.max(),lat.min(),lat.max()),cmap='jet',origin='upper',vmin=-12,vmax=12)

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('Cold nights')
#plt.savefig("../Assignment-2/plots/fig2/cold nights spatial")
```

```python
cold_nights_ts = xr.open_dataset('cold_days_timeseries.nc')
cold_nights_ts = cold_nights_ts['very_cold_days_percent_wrt_10th_percentile_of_reference_period'][:]
cold_nights_ts = cold_nights_ts.to_numpy()
cold_nights_ts = cold_nights_ts[:,0,0]
cold_nights_ts = list(cold_nights_ts)
cold_nights_ts.pop(np.argmin(cold_nights_ts))
```

```python
year = []
for i in range (1960,2020):
    year.append(i)
year.remove(1990)
```

```python
plt.figure(figsize=(8,5))
plt.bar(year,cold_nights_ts)
p = np.polyfit(year,cold_nights_ts,21)
trendline = np.polyval(p,year)
plt.plot(year,trendline)
plt.title("Cold nights")
#plt.savefig("../Assignment-2/plots/fig2/cold nights time series")
```
