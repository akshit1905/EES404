```python
import numpy as np
import xarray as xr

import matplotlib.pyplot as plt
import cartopy as cr
import cartopy.feature as cf
import cartopy.crs as ccrs
```

```python
file = xr.open_dataset('skt.sfc.mon.mean.nc')
```

```python
data_may = file.sel(time=file.time.dt.month.isin([5]))
data_june = file.sel(time=file.time.dt.month.isin([6]))
data_july = file.sel(time=file.time.dt.month.isin([7]))
data_aug = file.sel(time=file.time.dt.month.isin([8]))
data_sep= file.sel(time=file.time.dt.month.isin([9]))
data_JJAS = file.sel(time=file.time.dt.month.isin([6,7,8,9]))
```

```python
data_may_mean = np.mean(data_may['skt'],axis=(0))
data_june_mean = np.mean(data_june['skt'],axis=(0))
data_july_mean = np.mean(data_july['skt'],axis=(0))
data_aug_mean = np.mean(data_aug['skt'],axis=(0))
data_sep_mean = np.mean(data_sep['skt'],axis=(0))
data_JJAS_mean = np.mean(data_JJAS['skt'],axis=(0))
```

```python
lon = file.lon; lat = file.lat
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(data_may_mean,extent=(lon.min(),lon.max(),lat.min(),lat.max()),cmap='jet',origin='upper')

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3,label='skt')

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False
plt.title('Surface Temp. Climatology for May')
plt.savefig('../summer monsoon/task1_plots/Surface Temp. Climatology for May.jpeg')
```
