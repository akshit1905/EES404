```python
import pandas as pd
import numpy as np
import xarray as xr

import matplotlib.pyplot as plt
import cartopy as cr
import cartopy.feature as cf
import cartopy.crs as ccrs
```

```python
month = ['may','july','jjas']

for i in month:
    data = xr.open_dataset("./mslp."+i+".nc")
    data = data.mean(dim='time')
    data = data['mslp']

    lat = data.lat; lon = data.lon

    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))
    mp = ax.imshow(data/100,extent=(lon.min(),lon.max(),lat.min(),lat.max()),cmap='jet',origin='upper')

    ax.add_feature(cf.LAND)
    ax.add_feature(cf.COASTLINE)
    ax.add_feature(cf.OCEAN)

    cbar = fig.colorbar(mp, shrink=0.8,label='MSLP(hPa)')
    cbar.minorticks_on()

    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                      linewidth=2, color='gray', alpha=0.5, linestyle='--')
    gl.xlabels_top = False
    gl.ylabels_right = False
    plt.title('MSLP('+i.capitalize()+')' + ' '  + 'MSLP(hPa)')
    plt.savefig('../summer monsoon/task5_plots/MSLP('+i.capitalize()+')' + ' '  + 'MSLP(hPa)')

    
```
