```python
import numpy as np
import matplotlib.pyplot as plt

import xarray as xr

import cartopy as cr
import cartopy.crs as ccrs
import cartopy.feature as cf
```

```python
file_min = xr.open_dataset('../task-2/data/cru_ts4.06.1981.2020.tmn.MAM.dat.nc')
file_max = xr.open_dataset('../task-2/data/cru_ts4.06.1981.2020.tmx.MAM.dat.nc')
```

```python
file_min_avg = np.mean(file_min['tmn'],axis=(0))
file_max_avg = np.mean(file_max['tmx'],axis=(0))
```

```python
file_max_avg.shape
```

Maximum Temp

```python
lon_max = file_max_avg.lon
lat_max = file_max_avg.lat
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(file_max_avg,extent=(lon_max.min(),lon_max.max(),lat_max.min(),lat_max.max()),cmap='jet',origin='lower')

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3,label='Temperature (degC)')

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('Mean maximum temperature (degC)')
plt.savefig("../task-2/plots/Mean maximum temperature.jpeg")
```
