```python
import matplotlib.pyplot as plt
import numpy as np

import xarray as xr
```

ANNUAL

```python
global_file = xr.open_dataset('../task-1/data/air.mon.ltm.nc')
tropic_file = xr.open_dataset('../task-1/data/air.mon.ltm.tropic.nc')
northpole_file = xr.open_dataset('../task-1/data/air.mon.ltm.north_pole.nc')
india_file = xr.open_dataset('../task-1/data/air.mon.ltm.india.nc')
```

```python
global_avg = np.mean(global_file['air'],axis=(0,2,3))
tropic_avg = np.mean(tropic_file['air'],axis=(0,2,3))
northpole_avg = np.mean(northpole_file['air'],axis=(0,2,3))
india_avg = np.mean(india_file['air'],axis=(0,2,3))
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Annual vertical Thermal Structure')

plt.subplot(2,2,1)
plt.plot(global_avg,global_avg.level)
plt.ylim(1000,0)
plt.legend(['Global_mean'])
plt.ylabel('Pressure labels')

plt.subplot(2,2,2)
plt.plot(tropic_avg,tropic_avg.level)
plt.legend(['Tropic_mean'])
plt.ylim(1000,0)

plt.subplot(2,2,3)
plt.plot(northpole_avg,northpole_avg.level)

plt.legend(['NorthPole_mean'])
plt.ylim(1000,0)
plt.xlabel('Temperature')
plt.ylabel('Pressure labels')

plt.subplot(2,2,4)
plt.plot(india_avg,india_avg.level)
plt.legend(['India_mean'])
plt.ylim(1000,0)
plt.xlabel('Temperature')

plt.savefig("../task-1/plots/Annual vertical thermal structure.jpeg")
```

WINTER

```python
global_winter_file = xr.open_dataset('../task-1/data/air.mon.ltm.global_winter.nc')
tropic_winter_file = xr.open_dataset('../task-1/data/air.mon.ltm.tropic_winter.nc')
northpole_winter_file = xr.open_dataset('../task-1/data/air.mon.ltm.north_pole_winter.nc')
india_winter_file = xr.open_dataset('../task-1/data/air.mon.ltm.india_winter.nc')
```

```python
global_winter_avg = np.mean(global_winter_file['air'],axis=(0,2,3))
tropic_winter_avg = np.mean(tropic_winter_file['air'],axis=(0,2,3))
northpole_winter_avg = np.mean(northpole_winter_file['air'],axis=(0,2,3))
india_winter_avg = np.mean(india_winter_file['air'],axis=(0,2,3))
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Winter(DJF) vertical Thermal Structure')
plt.subplot(2,2,1)
plt.plot(global_winter_avg,global_winter_avg.level)
plt.ylim(1000,0)
plt.legend(['Global_mean'])
plt.ylabel('Pressure labels')

plt.subplot(2,2,2)
plt.plot(tropic_winter_avg,tropic_winter_avg.level)
plt.legend(['Tropic_mean'])
plt.ylim(1000,0)

plt.subplot(2,2,3)
plt.plot(northpole_winter_avg,northpole_winter_avg.level)

plt.legend(['NorthPole_mean'])
plt.ylim(1000,0)
plt.xlabel('Temperature')
plt.ylabel('Pressure labels')

plt.subplot(2,2,4)
plt.plot(india_winter_avg,india_winter_avg.level)
plt.legend(['India_mean'])
plt.ylim(1000,0)
plt.xlabel('Temperature')

plt.savefig("../task-1/plots/Winter(DJF) vertical thermal structure.jpeg")
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Vertical Thermal structural (Seasonal)')

plt.subplot(2,2,1)
plt.plot(global_winter_avg,global_winter_avg.level,'.-',label='Global')
plt.plot(tropic_winter_avg,tropic_winter_avg.level,'.-',label='Tropic')
plt.plot(northpole_winter_avg,northpole_winter_avg.level,'.-',label='North Pole')
plt.plot(india_winter_avg,india_winter_avg.level,'.-',label='India')
plt.legend()
plt.ylim(1000,0)
plt.grid(True)
plt.xlabel('Temperature')
plt.title('Winter')

plt.subplot(2,2,2)
plt.plot(global_summer_avg,global_summer_avg.level,'.-',label='Global')
plt.plot(tropic_summer_avg,tropic_summer_avg.level,'.-',label='Tropic')
plt.plot(northpole_summer_avg,northpole_summer_avg.level,'.-',label='North Pole')
plt.plot(india_summer_avg,india_summer_avg.level,'.-',label='India')
plt.legend()
plt.grid(True)
plt.ylim(1000,0)
plt.xlabel('Temperature')
plt.title("Summer")

plt.subplot(2,2,3)
plt.plot(global_avg,global_avg.level,'.-',label='Global')
plt.plot(tropic_avg,tropic_avg.level,'.-',label='Tropic')
plt.plot(northpole_avg,northpole_avg.level,'.-',label='North Pole')
plt.plot(india_avg,india_avg.level,'.-',label='India')
plt.legend()
plt.grid(True)
plt.ylim(1000,0)
plt.xlabel('Temperature')
plt.title('Annual')

plt.savefig("../task-1/plots/Vertical thermal structure (Seasonal).jpeg")
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Vertical Thermal structural')

plt.subplot(2,2,1)
plt.plot(global_avg,global_avg.level,'.-',label='Annual')
plt.plot(global_summer_avg,global_summer_avg.level,'.-',label='Summer')
plt.plot(global_winter_avg,global_winter_avg.level,'.-',label='Winter')

plt.legend()
plt.ylim(1000,0)
plt.grid(True)
plt.title('Global')
plt.xlabel('Temperature')
plt.ylabel('Pressure(hPa)')

plt.subplot(2,2,2)
plt.plot(tropic_avg,tropic_avg.level,'.-',label='Annual')
plt.plot(tropic_summer_avg,tropic_summer_avg.level,'.-',label='Summer')
plt.plot(tropic_winter_avg,tropic_winter_avg.level,'.-',label='Winter')

plt.legend()
plt.ylim(1000,0)
plt.grid(True)
plt.title('Tropical')
plt.xlabel('Temperature')
plt.ylabel('Pressure(hPa)')

plt.subplot(2,2,3)
plt.plot(northpole_avg,northpole_avg.level,'.-',label='Annual')
plt.plot(northpole_summer_avg,northpole_summer_avg.level,'.-',label='Summer')
plt.plot(northpole_winter_avg,northpole_winter_avg.level,'.-',label='Winter')

plt.legend()
plt.ylim(1000,0)
plt.grid(True)
plt.title('North Pole')
plt.xlabel('Temperature')
plt.ylabel('Pressure(hPa)')

plt.subplot(2,2,4)
plt.plot(india_avg,india_avg.level,'.-',label='Annual')
plt.plot(india_summer_avg,india_summer_avg.level,'.-',label='Summer')
plt.plot(india_winter_avg,india_winter_avg.level,'.-',label='Winter')

plt.legend()
plt.ylim(1000,0)
plt.grid(True)
plt.title('India')
plt.xlabel('Temperature')
plt.ylabel('Pressure(hPa)')

plt.savefig("../task-1/plots/Vertical thermal structure (Location).jpeg")
```
