```python
import numpy as np
import matplotlib.pyplot as plt

import xarray as xr

import cartopy as cr
import cartopy.crs as ccrs
import cartopy.feature as cf
```

```python
file = xr.open_dataset('../task-6/corr_pre_olr.nc')
file_pre = xr.open_dataset("../task-6/cru_ts4.06.1981.2010.JJA.mean.pre.dat.nc")
file_olr = xr.open_dataset("../task-6/olr.mon.ltm.1981-2010.regrid.nc")
```

```python
file['pre_olr_corr'] = np.mean(file['pre'],axis=(0))
file_pre['pre_olr_corr'] = np.mean(file_pre['pre'],axis=(0))
file_olr['pre_olr_corr'] = np.mean(file_olr['olr'],axis=(0))
```

```python
lon = file.lon; lat = file.lat
lon_pre = file_pre.lon; lat_pre = file_pre.lat
lon_olr = file_olr.lon; lat_olr = file_olr.lat
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(file['pre_olr_corr'],extent=(lon.min(),lon.max(),lat.min(),lat.max()),cmap='jet',origin='lower')

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('Spatial map correlation Pre/OLR')
plt.savefig("../task-6/plots/Spatial map correlation Pre_OLR.jpeg")
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(file_pre['pre_olr_corr'],extent=(lon_pre.min(),lon_pre.max(),lat_pre.min(),lat_pre.max()),cmap='jet',origin='lower',vmax=200,vmin=0)

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('Pre')
#plt.savefig("../task-2/plots/Mean maximum temperature.jpeg")
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(file_olr['pre_olr_corr'],extent=(lon_olr.min(),lon_olr.max(),lat_olr.min(),lat_olr.max()),cmap='jet',origin='lower')

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('OLR')
#plt.savefig("../task-2/plots/Mean maximum temperature.jpeg")
```
