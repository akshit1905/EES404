```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

import xarray as xr
```

```python
file = xr.open_dataset("../task-7/cru_tmp_mam_mean_CIR_std_fldmean_n.nc",decode_times=False)
```

```python
file['tmp'][:,0,0].shape
```

```python
year = []
for i in range (1901,2022):
    year.append(i)
```

```python
plt.figure(figsize=(10,5))
plt.plot(year,file['tmp'][:,0,0],'-o')

slope, intercept, r, p, std_err = stats.linregress(year,file['tmp'][:,0,0])
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, year))
plt.plot(year,mymodel,label= f'fit: y = {np.round(slope,2)}x + {np.round(intercept,2)}')

#plt.xlim(1990,2010)
plt.grid(True)
plt.legend()
plt.title('Standardized Temperature Anamoly (1901-2021)')
plt.xlabel('Year')
plt.savefig("../task-7/plots/Standardized Temperature Anamoly (1901-2021).png")
```
