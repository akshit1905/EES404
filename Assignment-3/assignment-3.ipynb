```python
import pandas as pd
import numpy as np
import xarray as xr

import matplotlib.pyplot as plt
import cartopy as cr
import cartopy.feature as cf
import cartopy.crs as ccrs
```

Spatial distribution of precipitable water  over the Indian region

```python
pre_wat = xr.open_dataset('output-pre-wat-india.nc')
#pre_wat
```

```python
pre_wat = np.mean(pre_wat['tcrw'],axis=(0))
```

```python
lon = pre_wat.longitude; lat = pre_wat.latitude

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

mp = ax.imshow(pre_wat,extent=(lon.min(),lon.max(),lat.min(),lat.max()),cmap='jet',origin='upper')

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3,label='kg/m^2')

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('Spatial distribution of precipitable water over the Indian region')
#plt.savefig("./plots/precipitable_water.png")
```

Spatial distribution of winds at 850 hPa and 200 hPa (vectors) over 60-100E & 0 - 35N region

```python
vwind = xr.open_dataset('output-vwind.india.nc')
uwind = xr.open_dataset('output-uwind.india.nc')
```

850 hPa

```python
vwind = vwind.sel(level=850); uwind = uwind.sel(level=850)
```

```python
vwind = np.mean(vwind['v'],axis=(0)); uwind = np.mean(uwind['u'],axis=(0))
```

```python
wind = np.sqrt(vwind**2 + uwind**2)
```

```python
lon = wind.longitude; lat = wind.latitude
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=0.0, globe=None))

skip = (slice(None, None, 10), slice(None, None, 10))
lon_2d, lat_2d = np.meshgrid(lon, lat)
ax.quiver(lon_2d[skip], lat_2d[skip], wind[skip], wind[skip], wind[skip],transform=ccrs.PlateCarree(), cmap='coolwarm', scale=300)

mp = ax.imshow(wind,extent=(lon.min(),lon.max(),lat.min(),lat.max()),cmap='jet',origin='upper')

ax.add_feature(cf.LAND)
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)

cbar = fig.colorbar(mp, shrink=0.3,label='m/s')

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

plt.title('Spatial distribution of winds at 850 hPa (vectors) over 60-100E & 0 - 35N region')
#plt.savefig("./plots/winds_850.png")
```
